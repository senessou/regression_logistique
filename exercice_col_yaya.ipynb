```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
#charger le jeu de données
data = pd.read_excel("Copie de Coeur.xlsx")
```

```python
df = data.copy()
df
```

```python
#verifier si des doublons si une ligne se reperte
df.duplicated().value_counts()
```

```python
#verifier si les données manquantes 
df.isna().value_counts()
```

```python
#supprimer une donnée manquant
df.dropna()
```

```python
#savoir comment supprimer les doublons
df.drop_duplicates()
```

```python
#verifier si pas de constante (les variables qui n'ont pas de variables de modalité) supprimer
df.nunique()
```

```python
valeur_qualitative =  df.select_dtypes('object').columns
valeur_qualitative
```

```python
del df["CŒUR"]
```

```python
valeur_quantitative =  df.select_dtypes(int, float).columns
valeur_quantitative
```

```python
#recoder les variables qualitatives 
def recoder(df):
    for col in valeur_qualitative:
        df[col] = df[col].astype("category").cat.codes
    return df  

recoder(df)
```

```python
#normaliser les variables quantitatives chaque valeur/ valeurs max
def normal(df):
    df_copy = df.copy()
    for col in valeur_quantitative:
        df_copy[col] = df_copy[col] / df_copy[col].abs().max()
    return df_copy
normal(df)
```

```python
#creer un dataframe x qui va contenir de age a pente et y coeur
x = df.iloc[:,:11]
x
```

```python
y = data.iloc[:,11:]
y
```

```python
#train_test_split() fonction on va lui donner en parametre x et y et le taux 0.3

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
#import la classe regresiion logistique  lineaire_model
```

```python
#(regression logistique avec lineaire_model)

#instanciation du modèle
modele_regLog = linear_model.LogisticRegression(random_state=0, solver = 'liblinear', multi_class = 'auto')

#training
modele_regLog.fit(x_train, y_train)

#prediction
prediction = modele_regLog.predict(x_test)
print('prédiction du modèle:', prediction)

#calcule de précision
precision = modele_regLog.score(x_test, prediction)
print('la précision du modèle', precision * 100)
```
